break:
Pernyataan break digunakan dalam loop (seperti for atau while) untuk menghentikan loop secara tiba-tiba.
Ketika break dieksekusi, loop akan langsung berhenti, dan program akan melanjutkan ke pernyataan setelah loop.
Ini berguna ketika Anda ingin menghentikan loop jika suatu kondisi tertentu terpenuhi.
Contoh penggunaan break:

python
Copy code
for i in range(1, 11):
    if i == 5:
        break  # Loop akan berhenti saat i sama dengan 5.
    print(i)
continue:
Pernyataan continue digunakan dalam loop untuk melompati iterasi saat ini dan melanjutkan ke iterasi berikutnya.
Ketika continue dieksekusi, pernyataan yang ada setelahnya dalam loop akan dilewati, dan loop akan melanjutkan ke iterasi berikutnya.
Ini berguna ketika Anda ingin melewati beberapa iterasi dalam loop berdasarkan kondisi tertentu tanpa menghentikan loop sepenuhnya.
Contoh penggunaan continue:

python
Copy code
for i in range(1, 11):
    if i % 2 == 0:
        continue  # Melompati angka genap.
    print(i)
Dengan demikian, break digunakan untuk menghentikan loop secara total, sedangkan continue digunakan untuk melompati iterasi tertentu dan melanjutkan ke iterasi berikutnya.
pass:
Pernyataan pass adalah pernyataan yang tidak melakukan apa-apa. Ini digunakan ketika Anda perlu menulis sintaksis yang sah, tetapi tidak ingin mengeksekusi kode apapun.
Biasanya digunakan sebagai placeholder sementara, ketika Anda merencanakan untuk mengisi kode lebih lanjut nanti.
Contoh penggunaan pass:

python
Copy code
for i in range(1, 11):
    if i % 2 == 0:
        pass  # Tidak melakukan apa-apa saat angka genap ditemukan.
    else:
        print(i)  # Mencetak angka ganjil.
Jadi, pass digunakan ketika Anda ingin memiliki suatu pernyataan di dalam blok kode, tetapi tidak ingin menjalankan operasi apa pun saat itu. Ini berbeda dari break dan continue, yang memiliki peran yang berbeda dalam mengontrol alur loop.

*

Basically number of lines it prints is equal to that number. (Hint: you need to use two for loops)

```python
def print_pattern(n):
    for i in range(1, n + 1):
        for j in range(i):
            print("*", end="")
        print()  # Print a new line after each row

# Test the function with different inputs
n=int(input("masukkan n : "))
print_pattern(n)

```
