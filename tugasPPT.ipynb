```python
#1
import math

print("masukkan tiga akngka terakhir nim anda : ")
a=int(input("a : "))
b=int(input("b : "))
c=int(input("c : "))

diskriminan = b**2-4*a*c
if a==0:
  print("bukan merupakan persamaa kuadrat")
else :
  if diskriminan > 0:
      x1 = float((-b + math.sqrt(diskriminan)) / (2*a))
      x2 = float((-b - math.sqrt(diskriminan)) / (2*a))
      print("Persamaan kuadrat memiliki dua akar real:", x1, "dan", x2)
  elif diskriminan == 0:
      x = float(-b / (2*a))
      print("Persamaan kuadrat memiliki satu akar real:", x)
  else:
      bil_real = float(-b / (2*a))
      bil_imajiner = float(math.sqrt(-diskriminan) / (2*a))
      print("Persamaan kuadrat memiliki dua akar kompleks:")
      print("Akar pertama:", bil_real, "+", bil_imajiner, "i")
      print("Akar kedua:", bil_real, "-", bil_imajiner, "i")
```

Ada beberapa data buah, dengan komponen : nama, kualitas dan harga.
Jenis tersebut: jeruk dengan harga 1000, mangga 2000, dan apel 3000, masing-masing
untuk kualitas 2.
Jika jeruk kualitas 1, maka harganya menjadi 1,5 kali lipat, dan kualitas 3 menjadi 75%
dari harga kualitas 2.
Mangga ada 2 kualitas, 1 dan 2 dengan harga kualitas 1 adalah 2 kali lipat dari kualitas 2.
Sedangkan apel hanya ada kualitas 2.
Masukan program adalah uang, nama buah, dan kualitas.
Luaran program adalah nama buah, kualitas dan harga.
Misal jika inputannya adalah sbb: untuk masukan ’5000’ (untuk uang), jeruk (nama), 3
(untuk kualitas),
maka akan didapatkan keluaran :’Anda membeli jeruk kualitas 3 dengan harga 750, dan
mendapatkan 6 buah jeruk dengan sisa uang 500

```python
def hitung_harga_buah(uang, nama_buah, kualitas) :
  harga=0
  jumlah_buah=0

  if nama_buah == 'jeruk':
      if kualitas == '1':
        harga=1000*1.5
      elif kualitas =='2':
        harga=1000
      elif kualitas =='3':
        harga=1000*0.75
  if nama_buah=='mangga':
      if kualitas=='1':
        harga=2000*2
      elif kualitas=='2':
        harga=2000
  if nama_buah=='apel' :
      if kualitas=='1':
        harga=3000

  jumlah_buah=int(uang//harga)
  return(jumlah_buah,uang-(jumlah_buah*harga))

uang =int(input("masukkan jumlah uang : "))
nama_buah=input("masukkan buah yang dipilih (jeruk/mangga/apel) : ")
kualitas=input("kualitas yang diinginkan (1/2/3) : ")

jumlah_buah, sisa_uang = hitung_harga_buah(uang, nama_buah, kualitas)


print("maka dengan banyaknya uang sejumlah", uang, "dan dengan pilihan kualitas ",kualitas, "anda menginginkan buah", nama_buah, "sebanyak",jumlah_buah, "dan mendapatkan sisa  uang sebanyak", sisa_uang)


```

2. Buatlah program untuk menentukan kelompok usia seseorang yang diinput o
pengguna, dengan ketentuan sebagai berikut :
a) Jika usia 5 tahun ke bawah (usia<=5), maka dia adalah balita.
b) Jika usia di atas 5 sampai 12 tahun , maka dia adalah anak-anak.
c) Jka usia di atas 12 tahun dan usia di bawah 18 tahun maka dia adalah remaja.
d) Jika usia di atas 18 tahun, maka dia adalah dewasa.
TUGAS PENCABANGAN

```python
#soal waktu parkir
def waktu_parkir():
  biaya=0
  kurang_dari_2jam =6000
  biaya_perjam=2500

  lama_parkir=int(input("masukkan waktu parkir (dalam jam) : "))
  if lama_parkir <= 2 :
    biaya = kurang_dari_2jam
  if lama_parkir > 2 :
    biaya= lama_parkir*biaya_perjam
  print("dengan total waktu parkir", lama_parkir, "jam, anda harus membayar sebanyak", biaya)

waktu_parkir()
```

```python
def faktorial(n) :
  hasil=1
  for i in range(1, n+1):
    hasil *= i
  return hasil

angka=int(input("masukkan angka : "))
nilai_faktorial=faktorial(angka)
print("maka nilai faktorial dari", angka, "adalah", nilai_faktorial)
```
