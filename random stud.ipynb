```python
def is_prima(angka):
    if angka <= 1:
        return False
    for x in range(2, int(angka ** 0.5) + 1):
        if angka % x == 0:
            return False
    return True

awal = int(input("Batas awal: "))
akhir = int(input("Batas akhir: "))

def cari_bil_prima(awal, akhir):
    list_bil_prima = []
    for x in range(awal, akhir + 1):
        if is_prima(x):
            list_bil_prima.append(x)
    return list_bil_prima

hasil_cari = cari_bil_prima(awal, akhir)
print("Bilangan prima antara", awal, "dan", akhir, "adalah:", hasil_cari)

```

```python
def hitung_harga_buah(uang, nama_buah, kualitas):
    harga = 0
    jumlah_buah = 0

    if nama_buah == 'jeruk':
        if kualitas == 1:
            harga = 1.5 * 1000
        elif kualitas == 2:
            harga = 1000
        elif kualitas == 3:
            harga = 0.75 * 1000
    elif nama_buah == 'mangga':
        if kualitas == 1:
            harga = 2 * 2000
        elif kualitas == 2:
            harga = 2000
    elif nama_buah == 'apel':
        harga = 3000

    jumlah_buah = int(uang // harga)

    return (jumlah_buah, uang - (jumlah_buah * harga))

uang = float(input("Masukkan jumlah uang: "))
nama_buah = input("Masukkan nama buah (jeruk/mangga/apel): ")
kualitas = int(input("Masukkan kualitas buah (1/2/3): "))

jumlah_buah, sisa_uang = hitung_harga_buah(uang, nama_buah, kualitas)

print("Anda membeli", jumlah_buah, nama_buah, "kualitas", kualitas, "dengan sisa uang", sisa_uang)
```

```python
import math

# Masukkan tiga angka terakhir NIM Anda
nim_last_three_digits = int(input("Masukkan tiga angka terakhir NIM Anda: "))

# Menggunakan tiga angka terakhir NIM untuk nilai a, b, dan c
a = nim_last_three_digits
b = nim_last_three_digits
c = nim_last_three_digits

# Menghitung diskriminan
diskriminan = b**2 - 4*a*c

# Mengecek apakah persamaan memiliki akar real atau kompleks
if diskriminan > 0:
    x1 = (-b + math.sqrt(diskriminan)) / (2*a)
    x2 = (-b - math.sqrt(diskriminan)) / (2*a)
    print("Persamaan kuadrat memiliki dua akar real:", x1, "dan", x2)
elif diskriminan == 0:
    x = -b / (2*a)
    print("Persamaan kuadrat memiliki satu akar real:", x)
else:
    real_part = -b / (2*a)
    imaginary_part = math.sqrt(-diskriminan) / (2*a)
    print("Persamaan kuadrat memiliki dua akar kompleks:")
    print("Akar pertama:", real_part, "+", imaginary_part, "i")
    print("Akar kedua:", real_part, "-", imaginary_part, "i")
```

Latihan 1: Kalkulator Sederhana
Buatlah kalkulator sederhana yang meminta dua angka dari pengguna dan operasi yang ingin dilakukan (tambah, kurang, kali, bagi), kemudian tampilkan hasilnya.

TENTANG RETURN DAN PASS

Return: "Return" digunakan dalam fungsi Python untuk mengembalikan nilai dari fungsi tersebut. Saat fungsi dipanggil dan mencapai pernyataan "return", eksekusi fungsi berhenti, dan nilai yang diikembalikan akan dikirim kembali ke tempat di mana fungsi itu dipanggil. Ini memungkinkan Anda mendapatkan hasil atau nilai dari fungsi untuk digunakan dalam bagian lain dari program.

Pass: "Pass" digunakan sebagai tanda placeholder atau penanda bahwa suatu blok kode seharusnya ada di sana, tetapi tidak ada tindakan yang perlu dilakukan dalamnya. Ini berguna ketika Anda ingin membuat kerangka kode atau fungsi yang belum diimplementasikan sepenuhnya. "Pass" memastikan kode tetap valid tanpa melakukan tindakan apa pun.

Singkatnya, "return" mengembalikan nilai dari fungsi, sedangkan "pass" hanya menandai tempat di mana kode akan ada nanti tetapi tidak melakukan apa-apa pada saat ini.

```python
#bikin kalkulator sederhana

def tambah(a,b) :
  return a+b

def kurang(a,b) :
  return a-b

def kali(a,b) :
  return a*b

def bagi(a,b) :
  if b == 0 :
    print("tidak dapat dibagi dengan 0")
  else :
    return a//b


pilihan=int(input("1. tambah \n2.kurang \n3.kali \n4.bagi \nmasukkan pilihan anda diantara ketiganya(1/2/3/4) : "))
x=int(input("masukkan angka pertama : "))
y=int(input("masukkan angka kedua : "))
if pilihan == 1 :
  print(x, "+", y, "=", tambah(x,y))
elif pilihan ==2 :
  print(x, "-", y, "=", kurang(x,y))
elif pilihan == 3 :
  print(x, "*", y, "=", kali(x,y))
elif pilihan== 4 :
  print(x, "//", y, "=", bagi(x,y))



```
